```python
import json
from pandas.io.json import json_normalize
```

```python
#Q2: Find the top 10 major project themes (using column 'mjtheme_namecode')

# load json as string
world_data = json.load((open('world_bank_projects.json')))
```

```python
# use normalization to create tables from nested element
theme_project_data = json_normalize(world_data, 'mjtheme_namecode')
theme_project_data
```

```python
# Group by name
theme_project_data_grp = theme_project_data.groupby("name")
theme_project_data_grp
```

```python
# Get the count of projects per country in descending order
top_themes_with_projects = theme_project_data_grp.size().sort_values(ascending=False)
top_themes_with_projects
```
