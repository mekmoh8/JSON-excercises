```python
import json
from pandas.io.json import json_normalize
import pandas as pd
import numpy as np
```

```python
#Q3
#You will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.

# load json as string
world_data = json.load((open('world_bank_projects.json')))

# use normalization to create tables from nested element
theme_project_data = json_normalize(world_data, 'mjtheme_namecode')
theme_project_data
```

```python
# Removing duplicates 
theme_prj_data_no_dup = theme_project_data.drop_duplicates()
theme_prj_data_no_dup
```

```python
# Filling spaces as NaN
theme_prj_data_no_dup = theme_prj_data_no_dup[theme_prj_data_no_dup != '']
theme_prj_data_no_dup
```

```python
# Remove rows with NaN
clean_data = theme_prj_data_no_dup.dropna()
clean_data
```

```python
# We change 'code' to be an integer instead of a string so we can sort
clean_data1 = clean_data[['code']].astype(int)
clean_data2 = clean_data[['name']]
```

```python
# Combine clean_data1 & clean_data2 using concat()
final_data = pd.concat([clean_data1, clean_data2], axis=1)
final_data
```

```python
# Sort using sort_values() and reset the index
final_data = pd.concat([clean_data1, clean_data2], axis=1)
final_data = final_data.sort_values('code')
final_data = final_data.reset_index(drop=True)
final_data
```
